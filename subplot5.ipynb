```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
df = sns.load_dataset('planets')
df
```

```python
plt.bar(df.distance,df.mass)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.xlabel('distance')
plt.ylabel('mass')
plt.title('Graph of planets')
plt.show()
```

```python
plt.plot(df.mass,df.distance)
```

```python
plt.boxplot(df.year,vert=False);
```

```python
plt.scatter(df.mass,df.distance)
```

```python
plt.barh(df.orbital_period,df.mass)
```

```python
plt.hist(df.distance)
# plt.show()
```

```python
name = ['banana','mango','orange','cherry']
price = [200,500,150,150]
plt.pie(price,labels=name)
plt.show()
```

```python
fig = plt.figure(figsize =(6, 3))
y = np.array([40,20,25,15])
mylabels=['Apple','Banana','Cherry','Mango']
plt.pie(y,labels=mylabels,shadow=True,explode=[0,.2,.3,.1])
plt.show()
plt.hist(y)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(221)
plt.bar(df.mass,df.year,color='cyan')
plt.title('Planet mass vs year')

plt.subplot(222)
plt.boxplot(df.year)
plt.title('Boxplot')

plt.subplot(223)
plt.barh(df.mass,df.distance,color='g')
plt.title('horizontal plot')

plt.subplot(224)
plt.hist(df.year,color='red')
plt.title('histogram')
plt.suptitle('subplot')

plt.tight_layout()
plt.show()
```
